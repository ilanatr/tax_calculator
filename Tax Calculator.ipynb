```python
#create function for monthly take_home pay after taxes and pensions etc.

#parameters = income, tax, if eligible 30%, tax minus 30% ruling, holiday pay

def take_home_pay(annual_income, thirty_percent):
    #calculate tax first 
    if annual_income < 73031:
        tax_rate = 0.3693
    else:
        tax_rate = 0.495

    income_tax = annual_income*tax_rate

    #apply 30% ruling
    if thirty_percent == 'yes':
     income_tax = annual_income*0.30 # Subtract 30% of income from tax
     
    #calculate tax first
    annual_take_home = annual_income - income_tax
    monthly_take_home = (annual_income - income_tax)/12
    print('Annual take home:', annual_take_home)
    print('Monthly take home:', monthly_take_home)

# 36.93%  < 70; 49.5% 
```
